```python
import pandas as pd
import numpy as np
from pandarallel import pandarallel
pandarallel.initialize()
```

```python
tracks_df = pd.read_csv('../data/tracks.csv.zip')
```

```python
display(tracks_df.shape)
tracks_df.release_date = tracks_df.release_date.astype('datetime64')
display(tracks_df.head())
display(tracks_df.release_date.describe())
```

```python
tracks_df.release_date.iloc[0].year
tracks_df.artists = tracks_df.artists.apply(lambda x: ','.join(eval(x)))
tracks_df.id_artists = tracks_df.id_artists.apply(lambda x: ','.join(eval(x)))
```

```python
display(tracks_df.isna().sum())

# Filter out  rows with null values for the 'name' column 
tracks_df = tracks_df[tracks_df.name.isna() == False]

display(tracks_df.isna().sum())

```

```python
match_artist = lambda x: [x in artists.split(',') for artists in tracks_df.artists]
match_song = lambda x: [len(set(x.lower().split()).intersection(set(sng.lower().split()))) >= len(x.split())  for sng in tracks_df.name]

display(tracks_df[match_artist('Bruno Mars')])
display(tracks_df[match_song("Don't worry be happy")])
```

```python
spotify_df = pd.read_csv('../data/spotify_songs.csv.zip')
display(spotify_df.shape)
spotify_df.head()
```

```python
display(spotify_df.isna().sum())
spotify_df = spotify_df[spotify_df.lyrics.isna() == False]
display(spotify_df.isna().sum())
```

```python
martist = lambda x: [x in artists for artists in spotify_df.track_artist]
msong = lambda x: [len(set(x.lower().split()).intersection(set(sng.lower().split()))) >= len(x.split())  for sng in spotify_df.track_name]
mlyric = lambda x: [len(set(x.lower().split()).intersection(set(sng.lower().split()))) >= len(x.split())  for sng in spotify_df.lyrics]
```

```python
display(spotify_df[martist('Chapman')])
display(spotify_df[msong("Don't worry be happy")])
display(spotify_df[mlyric("Michelle")])
```

```python
spotify_lyrics_df = spotify_df.filter(items=['track_id', 'lyrics']).rename(columns={'track_id':'id'})
display(spotify_lyrics_df.shape)
spotify_lyrics_df.head()
```

```python
display(tracks_df.shape)
lyrics_df = tracks_df.merge(spotify_lyrics_df, how='left', on='id')
display(lyrics_df.shape)
display(lyrics_df.lyrics.isna().sum())
display(lyrics_df[lyrics_df.lyrics.isna() == False].shape)
lyrics_df.head()
```

```python
display(lyrics_df.isna().sum())
lyrics_df.lyrics = lyrics_df.lyrics.apply(lambda x: '' if x is np.NaN else x)
display(lyrics_df.isna().sum())
```

```python
mlartist = lambda x: [x in artists for artists in lyrics_df.artists if artists]
mlsong = lambda x: [len(set(x.lower().split()).intersection(set(sng.lower().split()))) >= len(x.split())  for sng in lyrics_df.name ]
mllyric = lambda x: [len(set(x.lower().split()).intersection(set(sng.lower().split()))) >= len(x.split())  for sng in lyrics_df.lyrics]
```

```python
display(lyrics_df[mlartist('Tracy Chapman')])
display(lyrics_df[mlsong("don't worry be happy")])
display(lyrics_df[mllyric("Michelle")])
```
